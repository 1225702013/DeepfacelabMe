## （3）、训练模型
### 开始训练前，请配置模型运行参数 
> - 设置训练模型的类型 Model= "**` SAEHD `**" 可选的类型["SAEHD", "AMP", "Quick96", "XSeg"]；
> - 设置训练模型多少分钟自动保存 Saving_time= **` 30 `** 单位分钟为分钟，，取值范围（5-60分钟）设置为`0` ，则默认`25分钟`自动保存 ；
> - 设置多少分钟自动保存模型并结束训练 Train_time= **` 300 `** 单位分钟，取值范围（5-480分钟）设置为**` 0 `**则不会自动结束训练，需要手动结束训练  ；
> - 设置训练模型的类型 Auto_config = **` True `** 设置为**` True `**,用【`模型名_SAEHD_configuration_file.yaml` 】配置文件参数运行训练，设置为**` False `**就用【` 模型名_SAEHD_data.dat ` 】文件参数进行训练 ；


```python
# Train param set;
Model = "SAEHD"         #model param ["SAEHD", "AMP", "Quick96", "XSeg"]
Saving_time = 26        #param [5-60m]
Training_time = 0       #param [5-480m]

# Config param
Auto_config = True     #param [type:"boolean"]
Saved_models_type = "readpick-savepick" #Save model param ["readpick-savepick", "readjob-savejob"]

import datetime
import pytz
if Saving_time <=0:Saving_time = 25      #This value cannot be modified
if Training_time <=0:Training_time = 0   #This value cannot be modified
if Auto_config: Auto_gen_config="--auto-gen-config"
else:Auto_gen_config=" "
if Path("/kaggle/working/DeepFaceLab").exists() and Path("/kaggle/working/model").exists():
    start_time=datetime.datetime.now(pytz.timezone('PRC'))
    print('训练开始时间:'+start_time.strftime("%Y-%m-%d %H:%M"))
    %cd /kaggle/working/DeepFaceLab
    Train_cmd="main.py train --training-data-src-dir /kaggle/input/datasrc --training-data-dst-dir /kaggle/input/datadst --model-dir /kaggle/working/model --model "+str(Model)+" "+Auto_gen_config+" --saving-time "+str(Saving_time)+" --training-time "+str(Training_time)+" --saved-models-type "+Saved_models_type+" --no-preview --silent-start"
    !python $Train_cmd
    now_time = datetime.datetime.now(pytz.timezone('PRC'))
    print('训练结束，本次训练从['+start_time.strftime("%Y-%m-%d %H:%M")+']开始-至['+now_time.strftime("%Y-%m-%d %H:%M")+']')
    print('训练总时长:'+str(now_time -start_time))
else:
    print('运行出错，可能的出错原因：(1)未恢复运行环境；(2)未正确安装DeepfacelabMe程序；(3)没有model目录')
```

# （三）： 模型压缩打包
### 地址为working目录，默认打包文件名：【‘model+日期+时间’.zip】，可自行修改

> - 设置 Model_Auto_zip=`Fase` 不打包模型；

```python
Model_Auto_zip=1

import datetime
import pytz
from pathlib import Path

if Path("/kaggle/working/down").exists():
    %rm /kaggle/working/down/*.zip
    print("Clean kaggle/working/down/*.zip!")

if Model_Auto_zip:
    %cd /
    if not Path("/kaggle/working/down").exists():
        %mkdir /kaggle/working/down
    zip_filename=f"/kaggle/working/down/model_"+datetime.datetime.now(pytz.timezone('PRC')).strftime("%Y-%m-%d_%H%M")+".zip"
    zip_cmd="-r "+zip_filename+" /kaggle/working/model/*.*"     
    !zip $zip_cmd
    print("zip File Save to: "+ zip_filename +" 请在输出环境Data下自行下载! ")
    %cd /kaggle/working/down
    !tree -a
else:
    print("本次运行完毕，模型打包未开启")
```
